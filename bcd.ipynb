BREAST CANCER DETECTION USING ANN

```python
#importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#importing dataset
dataset=pd.read_csv('data.csv')
```

```python
#printing the dataset
dataset
```

```python
#Encoding the DIAGNOSIS Column into 0's and 1's
from sklearn.preprocessing import LabelEncoder 
encoder=LabelEncoder() 
dataset['diagnosis']=encoder.fit_transform(dataset['diagnosis']) 
dataset
```

```python
#dropping the unnecessary columns
dataset=dataset.drop(['Unnamed: 32','id'],axis=1) 
```

```python
#Description of dataset
dataset.describe()
```

```python
#information of dataset
dataset.info()
```

```python
#Heatmap showing the correlation values
plt.figure(figsize=(30,15))
sns.heatmap(dataset.corr(),annot=True)
```

```python
#pairplots showing the values of different attributes contributing towards the presence/absence of breast cancer
sns.pairplot(dataset,hue='diagnosis',vars=['radius_mean','texture_mean','perimeter_mean','area_mean'
,'smoothness_mean'])
```

```python
#getting independent and dependent values from the dataset
X=dataset.iloc[:,1:]
X
```

```python
#getting independent and dependent values from the dataset
y=dataset.iloc[:,0] 
y
```

```python
X=X.values
```

```python
y=y.values
```

```python
#Normalization (scaling all the values b/w 0 and 1)
from sklearn.preprocessing import MinMaxScaler 
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
#Splitting into training and test set
from sklearn.model_selection import train_test_split 
train_x,test_x,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=0)
test_x.shape
```

```python
#Applying ANN and finding the best values for the hyperparameters
import tensorflow
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier 
from sklearn.model_selection import GridSearchCV 
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 30)) 
    classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu')) 
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid')) 
    classifier.compile(optimizer ='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

classifier = KerasClassifier(build_fn = build_classifier)

parameters={'batch_size': [10,20,30], 'epochs' : [100,200,300]}
#now we will implement gridsearch
grid_search=GridSearchCV(estimator=classifier,param_grid=parameters,scoring='accuracy',cv=5,n_jobs=
-1)
grid_search=grid_search.fit(train_x,train_y) 
best_parameters=grid_search.best_params_ 

```

```python
best_parameters
```

```python
#Applying ANN with the optimized values of hyperparameters
from keras.models import Sequential 
from keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 30)) 
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu')) 
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid')) 
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) 
history=classifier.fit(train_x,train_y,epochs=100,batch_size=30,validation_data = (test_x,test_y))
```

```python
history.history.keys()
```

```python
#Training and validation accuracy
accuracy = history.history['accuracy'] 
val_accuracy = history.history['val_accuracy'] 
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(accuracy))

plt.plot(epochs, accuracy, 'ro', label='Training Accuracy') 
plt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy') 
plt.title('Training and Validation accuracy')
plt.legend()
```

```python
plt.plot(epochs, loss, 'ro', label='Training Loss') 
plt.plot(epochs, val_loss, 'b', label='Validation Loss') 
plt.title('Training and validation loss')
plt.legend() 
plt.show()
```

```python
#Predections on Test data
y_pred = classifier.predict(test_x)
#probability showing the individual having breast cancer (Malignant) 
# #probability over 50% would indicate the presence of breast cancer
y_pred = (y_pred > 0.5) #Converting the probabilities into 'true' and 'false'
```

```python
#Confusion matrix and heatmap for visualizing the accuracy of ANN on the test data
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y, y_pred) 
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
#Accuracy on test data
print("Accuracy Score:",round((cm[0,0]+cm[1,1])/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1])*100,2))
```
